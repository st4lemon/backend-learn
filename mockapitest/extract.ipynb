```python
import pandas as pd
import csv
import sklearn
from sklearn import datasets

from IPython.display import display

```

```python

def data_to_csv(data, name):
    df = pd.DataFrame(data['data'], columns=data.feature_names)
    df['target'] = pd.DataFrame(data['target']) \
        .apply(lambda x : (data['target_names'][x]) if 'target_names' in data else x)

    display(df)

    df.to_csv(f'data/skdata/{name}_data.csv')
```

```python

data_to_csv(datasets.load_iris(), 'iris')
data_to_csv(datasets.load_digits(), 'digits')
data_to_csv(datasets.load_diabetes(), 'diabetes')
data_to_csv(datasets.load_breast_cancer(), 'breast_cancer')
data_to_csv(datasets.load_wine(), 'wine')
```

```python
import requests
import json

```

```python

def post_csv(file: str):
    with open(f'data/skdata/{file}.csv', 'r') as f:
        requests.post(
            url = "http://localhost:8000/data/upload",
            files = {
                "file": (f"{file}.csv", f, "text/csv")
            },
            params = {
                'filename': file
            }
        )
```
